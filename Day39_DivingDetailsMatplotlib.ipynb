if you want to plot both the circles(dots) with the lines:
 you will first plot the line then the circle(dots)

```python
import numpy as np
import matplotlib.pyplot as plt

day=np.arange(1,8)
temperature=np.random.randint(20,31,7)
temperature2=np.random.randint(10,17,7)

fig,ax=plt.subplots()

ax.set(xlabel='Day',ylabel='Temperature')

ax.plot(day,temperature,color='r')  
ax.plot(day,temperature,'or',label='the highs')
ax.plot(day,temperature2,color='b')  
ax.plot(day,temperature2,'ob',label='the lows')
ax.grid(linestyle=":",color='green',linewidth=1.0)

plt.legend()

plt.show()
```

stackplots
these are plots that starck on top of each other in areas

```python
year=np.arange(2020,2025)
form1=np.random.randint(0,100,5)
form2=np.random.randint(0,92,5)
form3=np.random.randint(0,86,5)
form4=np.random.randint(0,96,5)

labels=['form 1','form 2','form 3','form 4']
plt.stackplot(year,form1,form2,form3,form4,labels=labels)
plt.legend()
plt.show()
```
