```python
import numpy as np
import matplotlib.pyplot as plt
#plt.ioff()
plt.plot([1, 2, 3, 4], [1, 4, 9, 16],'go')
plt.axis([0, 5, 0, 17])
plt.show()
```

the above data is a x,y pair in harmonised version

the form is [x][y]
where [x] holds all x values
and [y] holds all y values
the 'go' argument is for producing the dots of the plot the g stands for green color
the matlpolot lib hold colors for the dot in red,green,blue,yellow

the axis [1,2,1,10]
1,2 holds for the limits of the x axis

1,10 holds for the limits of y-axis
