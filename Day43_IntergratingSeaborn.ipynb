below is an example of intergrating matplotlib with seaborn styles

```python
'''Declaration first'''
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.set_style("darkgrid") # Apply Seaborn style
data = [1, 2, 3, 4, 5, 6, 7, 8, 9]
plt.plot(data, marker='o', linestyle='-', color='b')
# Matplotlib plot with Seaborn styling
plt.title("Example Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

the background above is proudly from seaborn styles

example 2 seeing how seaborn works easily with pandas data frame

first declaring data frame object


```python
data = {"Category": ["A", "B", "C", "D"], "Values": [10, 20, 15, 25]}
df = pd.DataFrame(data)
```

lets see how matplotlib works

```python
plt.bar((df["Category"]),(df["Values"]),color='m')
plt.title("Bar Plot of Categories")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

lets see how seaborn takes care of this

```python
sns.barplot(x="Category", y="Values", data=df,color='m')
plt.title("Bar Plot of Categories")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```
