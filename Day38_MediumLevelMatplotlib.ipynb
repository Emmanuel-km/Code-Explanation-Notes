```python
#importing stuffs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#generating x values
x=np.linspace(0,2,100)

#ploting specific values of:  
plt.plot(x,x,label='linear')
plt.plot(x,x**2+0.6657,label='quadratic')
plt.plot(x,x**3,label='cubic')
plt.plot(x,x*x,label='square')


'''normal grid '''
plt.grid()

''' enhanced grid'''
plt.grid(linestyle=":",color='#ff30c3',linewidth=1.0)

#labelling axis
plt.xlabel('x values')
plt.ylabel('x_worked values')

#giving the title
plt.title('plot of linear, quadratic, cubic\n, square values of varient x')

plt.legend()
plt.show()
```
