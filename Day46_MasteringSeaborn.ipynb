#### INTO SEABORN BASICS

A pretty interesting part about seaborn comes with in-built paramators/functions/methods

the set(),set.style(),despine() method/function

```python
import numpy as np
from matplotlib import pyplot as plt
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
sinplot()
plt.show()
```

with the set function

```python
import seaborn as sb
sb.set()
sinplot()
plt.show()
```

set style takes :  Darkgrid
 Whitegrid
 Dark
 White
 Ticks

**Despining**- removing axis top and right
{despine()}

```python
sb.set_style("white")
sinplot()
sb.despine()
plt.show()
```

### Scaling Seaborn plots

this can be done through the set_context()-----parameters{('paper','notebook','talk','poster')}

```python
sb.set_style("white")
sb.set_context('paper')
sinplot()
sb.despine()
plt.show()
```

```python
sb.set_style("white")
sb.set_context('notebook')
sinplot()
sb.despine()
plt.show()
```

```python
sb.set_style("white")
sb.set_context('talk')
sinplot()
sb.despine()
plt.show()
```

```python
sb.set_style("white")
sb.set_context('poster')
sinplot()
sb.despine()
plt.show()
```
