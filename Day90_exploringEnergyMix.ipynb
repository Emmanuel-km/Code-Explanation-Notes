```python
import seaborn as sns

import matplotlib.pyplot as plt

# Create dummy data
data = {'Category': ['Biofuels and Waste', 'Coal', 'Oil', 'Solar,wind and other renewable','Hydropower'], 'Values': [62.2,3.6, 16.7, 16.4,1]}

# Create pie chart
plt.figure(figsize=(8, 6),facecolor="#0f91258d")
colors = sns.color_palette("husl", len(data['Category']))
plt.pie(data['Values'], labels=data['Category'], autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('Energy mix ', fontsize=15, color='white')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6), facecolor="#0f91258d")
plt.pie(data['Values'], labels=data['Category'], autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('Energy mix in kenya ', fontsize=15, color='white')

# Change text color to white
for text in plt.gca().texts:
    text.set_color('white')

plt.tight_layout()
plt.show()
```

```python
import random

# generate random RGB colors for each category
colors_random = [(random.random(), random.random(), random.random()) for _ in data['Category']]

plt.figure(figsize=(8, 6), facecolor="#0f91258d")
plt.pie(data['Values'], labels=data['Category'], autopct='%1.1f%%', colors=colors_random, startangle=90, wedgeprops={'edgecolor': 'white'})
plt.title('Energy mix in Kenya', fontsize=15, color='white')

# Ensure text (labels and percentages) are visible on the dark background
for text in plt.gca().texts:
    text.set_color('white')

plt.tight_layout()
plt.savefig('energy_mix_pie_chart.png')
plt.show()
```
