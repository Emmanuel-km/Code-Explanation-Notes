Matplot is a module used to graph values and efficient for data visualisation

we will apply the knowledge of numpy and pandas for great data manipulation

```python
#first matplot figure

#importing stuffs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#generating x values
x=np.linspace(0,2,100)

#ploting specific values of:  
plt.plot(x,x,label='linear')
plt.plot(x,x**2+0.6657,label='quadratic')
plt.plot(x,x**3,label='cubic')
plt.plot(x,x*x,label='square')

#labelling axis
plt.xlabel('x values')
plt.ylabel('x_worked values')

#giving the title
plt.title('plot of linear, quadratic, cubic\n, square values of varient x')

plt.legend()
plt.show()
```

so the above plot represents data in its graphical simple visualisation

note linespace is favourate used because one can adjust the length of the value

see the sin graph with the code below

```python
#creating an array for x and y values in numpy
x=np.arange(0,10,0.2)
y=np.sin(x)

#matplotlib doing its magic
sin_graph,axis_toBePlotted=plt.subplots()
axis_toBePlotted.plot(x,y)

#after declaration methods
plt.title('a sine graph')
plt.show()
```

The subplot method holds a creteria that can make a plot of the same data divided into parts of the graph
